```python
# Drop unnecessary columns
cons_df = consumption_df.drop(["Data_Status","MSN"], axis = 1)
cons_df.head()
```

```python
# convert the data change columns to rows so our data is linear
cons_df = cons_df.melt(id_vars=["State"], 
        var_name="Year", 
        value_name="Total Consumed(Billion Btu)")
pd.set_option('display.max_rows', None)
cons_df.head()
```

```python
# select only GDPRX type
gdp_df = df[df["MSN"] == "GDPRX"]
gdp_df.head()
```

```python
# Drop unnecessary columns
gdp_df = gdp_df.drop(["Data_Status","MSN"], axis = 1)
gdp_df.head()
```

```python
# convert the data change columns to rows so our data is linear
gdp_df =gdp_df.melt(id_vars=["State"], 
        var_name="Year", 
        value_name="GDP(Million dollars)")
pd.set_option('display.max_rows', None)
gdp_df.head()
```

```python
# Select population type
pop_df = df[df["MSN"] == "TPOPP"]
pop_df.head()
```

```python
# Drop unnecessary columns
pop_df=pop_df.drop(["MSN", "Data_Status"], axis = 1)
pop_df.head()
```

```python
# convert the data change columns to rows so our data is linear
pop_df =pop_df.melt(id_vars=["State"], 
        var_name="Year", 
        value_name="Population(Thousand)")
pd.set_option('display.max_rows', None)
pop_df.head()
```

```python
# read the data for total energy average price
price_df = pd.read_csv("db/prices_all_states.csv")
price_df.head()
```

```python
# select the price type TETCD
price_df = price_df.loc[price_df["MSN"] == "TETCD"]
price_df.head()
```

```python
# remove US values
price_df = price_df.drop(price_df.index[price_df["State"]=="US"])
price_df.head()
```

```python
# Drop unnecessary columns
price_df=price_df.drop(["Data_Status", "MSN"], axis = 1)
price_df.head()
```

```python
# Replace state abbr. with state names
dict = {
    "AK": "Alaska", "AL":"Alabama", "AR": "Arkansas", "AZ": "Arizona", "CA": "California","CO": "Colorado", "CT": "Connecticut",
"DC": "District of Columbia","DE":"Delaware","FL":"Florida","GA": "Georgia",
"HI":"Hawaii","IA": "Iowa","ID":"Idaho","IL": "Illinois","IN":"Indiana",
"KS": "Kansas","KY" :"Kentucky","LA": "Louisiana","MA":"Massachusetts","MD":"Maryland",
"ME" :"Maine","MI":"Michigan","MN":"Minnesota","MO": "Missouri","MS" : "Mississippi","MT":"Montana",
"NC": "North Carolina","ND": "North Dakota","NE":"Nebraska","NH":"New Hampshire","NJ":"New Jersey","NM": "New Mexico",
"NV": "Nevada","NY": "New York","OH": "Ohio","OK": "Oklahoma","OR": "Oregon","PA":"Pennsylvania",
"RI":"Rhode Island","SC": "South Carolina","SD":"South Dakota","TN": "Tennessee","TX": "Texas",
"UT": "Utah","VA": "Virginia","VT":"Vermont","WA":"Washington","WI":"Wisconsin",
"WV": "West Virginia","WY":"Wyoming"
}
price_df["State"] = price_df["State"].map(dict)
price_df.head(10)
```

```python
# convert the data change columns to rows so our data is linear
price_df =price_df.melt(id_vars=["State"], 
        var_name="Year", 
        value_name="Energy Price(dollars per million btu)")
#pd.set_option('display.max_rows', None)
price_df.head()
```

```python
# Change data types for year values
cons_df["Year"]=cons_df["Year"].astype(str).astype(int)
cons_df.info()

pop_df["Year"]=pop_df["Year"].astype(str).astype(int)
pop_df["Year"]

gdp_df["Year"]=gdp_df["Year"].astype(str).astype(int)
gdp_df["Year"]

prod_df["Year"] = prod_df["Year"].astype(str).astype(int)
prod_df.info()

price_df["Year"] = price_df["Year"].astype(str).astype(int)
price_df.info()
```

```python
# Merge the dataframes

# merge renewable production and consumtion dataframes
states_df1 = pd.merge(prod_df, cons_df, on=["State", "Year"])

# merge population and gdp dataframes
states_df2 = pd.merge(gdp_df, pop_df, on=["State", "Year"])

# merge two dataframes
states_df3 = pd.merge(states_df1, states_df2, on=["State", "Year"])

# add prices dataframe
states_df = pd.merge(states_df3, price_df, on=["State", "Year"])

states_df.head()
```
